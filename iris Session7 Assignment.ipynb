```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
label=pd.DataFrame(list(map(lambda x :iris.target_names[x],iris.target)),columns=['Species'])
```

```python
iris = pd.concat([data,label],axis=1)
```

```python
print(iris.head())
```

## distplot()

```python
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.distplot( iris["sepal length (cm)"] , color="Green", ax=axes[0, 0])
sns.distplot( iris["sepal width (cm)"] , color="Red", ax=axes[0, 1])
sns.distplot( iris["petal length (cm)"] , color="Blue", ax=axes[1, 0])
sns.distplot( iris["petal width (cm)"] , color="Orange", ax=axes[1, 1])
plt.subplots_adjust(wspace=0, hspace=0)
sns.plt.show()
```

## boxplot()

```python
new_data = iris[["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]] 
print(new_data.head()) 
```

```python
plt.figure(figsize = (10, 7)) 
new_data.boxplot() 
```

##  countplot()

```python
plt.figure(figsize=(12,8))
sns.countplot(x="Species", data=iris,)
sns.plt.show()
```

## pairplot()

```python
sns.set_style("darkgrid");
sns.pairplot(iris, hue="Species", height=4);
plt.show()
```

## lmplot()

```python
sns.set_style("whitegrid");
sns.set(color_codes=True)
sns.lmplot(x='sepal length (cm)',y='petal length (cm)',hue='Species',data=iris)
plt.show()
```

## barplot()

```python
sns.set_style("whitegrid");
sns.barplot(x='Species',y='sepal length (cm)',data=iris)
plt.show()
```

## using heatmap, plot the correlation matrix

```python
corr = iris.corr()

sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns)
plt.show()
```
